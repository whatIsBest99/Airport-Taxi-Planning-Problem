```python
import os
import pandas
import codecs
import glob
import pandas as pd
```

```python
 def txtcombine():
    files = glob.glob('01/*/*.txt')
    
    all = codecs.open('all.txt','a')
    
    i = 0
    for filename in files:
        ++i
        print(str(i) + '\t' + filename)  
        with open(filename, encoding="utf8", errors='ignore') as f:
#             fopen=codecs.open(filename,'r',encoding='utf-8')  
            lines=[]  
            lines=f.readlines()
#             fopen.close()  
        for line in lines:    
            for x in line:  
                all.write(x)
       
#         #读取为DataFrame格式
#         al = pd.read_csv('all.txt', sep='|')
#         al.to_csv('all.csv')
```
